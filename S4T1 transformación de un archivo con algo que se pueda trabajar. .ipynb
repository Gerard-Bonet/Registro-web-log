```python
linearr= np.array ( lineas)# convertimos la lista en una array para más adelante vectorizar el cálculo y evitar iterar
```

```python
# cómo queremos separar todas las referencias por ";", vamos a eliminar todos los puntos y coma que ya existieran en el 
# documento, sustituyendo "; " por ","
def puntocoma(x):
    x= re.sub( ";",",",x)
    return x
coma =np. vectorize ( puntocoma)
linearr= coma(linearr)
print (linearr[113])  # tengo la línea 113 localizada con puntos y comas. 
```

```python
prueba1 = linearr[11234].copy()
# usaremos esta línea como campo de pruebas para luego aplicar sobre todo el bloque 
```

```python
# primero, substituímos los dos primeros espacios en blanco en una ";", de forma que quede www.web.com; 11.22.333.444;


prueba11= re.sub( "\s",";", prueba1,2 )

print (prueba11)


```

```python
# luego vamos a eliminar los dos primeros espacios en blanco que queden así como los dos guiones. 

prueba12= re.sub( "-","", prueba11,2)
prueba12=re.sub("\s","", prueba12,2)
print(prueba12)
# de esta manera nos quedan perfectamente separadas 3 variables por ";"
```

```python
# si observamos las siguientes  4 comillas, vemos que casualmente nos separan la variable tiempo del resultado del búsqueda, 
# así como el número de la variable X, y esta última del navegador. 
prueba13= re.sub('"', ";", prueba12,4)
prueba13= re.sub('"', "", prueba13,1)# luego eliminamos la siguiente comilla porque convertila en ";"nos crearía una variable
# más completámente vacían 
prueba13= re.sub('"', ";", prueba13,1)
print(prueba13)
```

```python
# por últimos nos quedaría separar el protocolo HTTP de la búsqueda. Que lo harémos por último
pru = linearr[113].copy()
def fun(x):
    x1=re.sub( "\s",";", x,2 )
    x2= re.sub( "-","", x1,2)
    x2=re.sub("\s","",x2,2)
    x3= re.sub('"', ";", x2,4)
    x3= re.sub('"', "", x3,1)
    x3= re.sub('"', ";", x3,1)
    return x3

fun2 =np. vectorize ( fun)
linearr= fun2(linearr)

print (linearr[22234])
```

```python
fichero = open ( "weblog.csv", "w")
fichero.write("Web;IP;Fecha;Busqueda;Numero;origen;Navegador;VLOG\n")
for i in range (0, len(linearr)):
    fichero.write(linearr[i])
fichero.close()
```
